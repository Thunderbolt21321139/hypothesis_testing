```python
from palmerpenguins import load_penguins
import numpy as np
import pandas as pd

penguins = load_penguins()
penguins.head()
```

```python
# remove penguins with no flipper length data

print(len(penguins))
penguins = penguins.dropna(subset=['flipper_length_mm'])
print(len(penguins))
```

```python
# remove penguins with sex data

print(len(penguins))
penguins = penguins.dropna(subset=['sex'])
print(len(penguins))
```

The null hypothesis is that male and female penguins have the same sized flippers. The alternate hypothesis is that male penguins have larger flippers. For this test we should use a right tailed t test.

```python
from scipy import stats

male_flipper_length = penguins[penguins['sex'] == 'male']['flipper_length_mm']
female_flipper_length = penguins[penguins['sex'] == 'female']['flipper_length_mm']

mean_male = male_flipper_length.mean()
mean_female = female_flipper_length.mean()

var_male = male_flipper_length.var(ddof=1)  # Sample variance
var_female = female_flipper_length.var(ddof=1)

n_male = len(male_flipper_length)
n_female = len(female_flipper_length)

t_statistic = (mean_male - mean_female) / ((var_male / n_male) + (var_female / n_female))**0.5
df = n_male + n_female - 2

# two-tailed test
#p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# left-tailed test
#p_value = stats.t.cdf(t_statistic, df)

# right-tailed test
p_value = 1 - stats.t.cdf(t_statistic, df)

print(f'Mean male: {mean_male}')
print(f'Mean female: {mean_female}')
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

Since the p value is less than the significance value of 0.05 we should reject the null hypothesis that male and female penguins have the same sized flippers.
